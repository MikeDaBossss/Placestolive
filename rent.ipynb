```
from bs4 import BeautifulSoup
import requests
import pandas as pd
seasons = ['spring','fall','summer','winter']
for season in seasons:
  spring = requests.get('https://www.currentresults.com/Weather/US/average-state-temperatures-in-'+season+'.php')
  spring = BeautifulSoup(spring.content, 'html.parser')
  table = spring.find_all('table', class_='articletable')
  table1 = table[0].find_all('td')
  table2 = table[1].find_all('td')
  table1list = []
  for counter,i in enumerate(table1):
    if counter%4 == 0:
      if counter == 0 :
        continue
      table1list.append([table1[counter-4].text,table1[counter-3].text,table1[counter-2].text,table1[counter-1].text])
  table1list
  table2list = []
  for counter,i in enumerate(table2):
    if counter%4 == 0:
      if counter == 0 :
        continue
      table2list.append([table2[counter-4].text,table2[counter-3].text,table2[counter-2].text,table2[counter-1].text])
  table2list
  weather = pd.DataFrame(table1list)
  weather1 = pd.DataFrame(table2list)
  weathertemp = pd.concat([weather,weather1])
  weathertemp.index = weathertemp[0]
  weathertemp.columns = [season,season,season,season]
  if season == 'spring':
    weatherfinal = pd.DataFrame()
    weatherfinal = weathertemp
  else:
    try:
      weatherfinal.drop(['key_0'],axis=1,inplace=True)
    except:
      None
    weatherfinal = pd.merge(weatherfinal,weathertemp,left_index=True,left_on=weatherfinal.index,right_on=weathertemp.index)
weatherfinal
```
